```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import spacy
import re
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bbc_data = pd.read_csv("./../bbc_news.csv")
```

```python
bbc_data.head()
```

```python
bbc_data.info()
```

```python
titles = pd.DataFrame(bbc_data["title"])
titles['lowercase'] = titles['title'].str.lower()
titles.head()
```

```python
en_stopwords = stopwords.words('english')
titles['no_stopwords'] = titles['lowercase'].apply(lambda x: ' '.join([word for word in x.split() if word not in (en_stopwords)]))
titles.head()
```

```python
titles['no_stopwords_no_punc'] = titles['no_stopwords'].apply(lambda x: re.sub(r"([^\w\s])", '', x))
titles.head()
```

```python
titles['tokens_clean'] = titles['no_stopwords_no_punc'].apply(lambda x: word_tokenize(x))
titles.head()
```

```python
lemmatizer = WordNetLemmatizer()
titles['tokens_clean_lemmatized'] = titles['tokens_clean'].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
titles.head()
```

```python
tokens_clean_lemmatized = [token for sublist in titles['tokens_clean_lemmatized'] for token in sublist]
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
spacy_doc = nlp(' '.join(tokens_clean_lemmatized))
```

```python
ner_df = pd.DataFrame([(e.text, e.label_) for e in spacy_doc.ents if e.label_], columns=['token', 'ner_tag'])
ner_df.head()
```

```python
ner_df_counts = ner_df.groupby(['token', 'ner_tag']).size().reset_index(name="counts").sort_values(by='counts', ascending=False)
ner_df_counts.head()
```
